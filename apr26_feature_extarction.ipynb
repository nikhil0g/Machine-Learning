```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('iris.csv',names=['sep_l','sel_w','pet_l','pet_w','label'])
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['label'] = le.fit_transform(dataset['label'])
```

```python
dataset
```

```python
X = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4].values
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X_reduced = pca.fit_transform(X)
```

```python
X_reduced
```

```python
X_reduced = pd.DataFrame(X_reduced,columns=['x1','x2'])
```

```python
X_reduced
```

```python
sns.scatterplot(x='x1',y='x2',hue=y,data=X_reduced)
```
